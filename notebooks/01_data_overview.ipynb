## Initial observations

The cleaned dataset contains 113,197 rows and 26 columns. There are 3 metadata columns: `Metadata_Plate`, `Metadata_Well`, and `Metadata_JCP2022`. The remaining columns include object counts and selected CellProfiler AreaShape features for cells and nuclei, as well as derived N/C ratio variables.

There are no missing values in the current cleaned dataset. The compound identifier column is `Metadata_JCP2022`, with 5 unique compound IDs in this file.

```python
df["Metadata_JCP2022"].unique()
compound_counts = (
    df["Metadata_JCP2022"]
    .value_counts()
    .rename_axis("Metadata_JCP2022")
    .reset_index(name="n_cells")
)

compound_counts
```
